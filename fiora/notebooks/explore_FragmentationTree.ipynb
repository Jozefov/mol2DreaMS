```python
import sys
from rdkit import Chem
from rdkit.Chem import Draw
from fiora.MOL.FragmentationTree import FragmentationTree
from fiora.MOL.mol_graph import mol_to_graph, get_adjacency_matrix, get_edges

```

```python
# Define the SMILES string for the molecule you want to fragment
smiles = "CC(C)C(CCCN(C)CCC1=CC(=C(C=C1)OC)OC)(C#N)C2=CC(=C(C=C2)OC)OC"  # Verapamil

# Create an RDKit molecule object
mol = Chem.MolFromSmiles(smiles)

# Generate the molecular graph using mol_to_graph function from mol_graph module
G = mol_to_graph(mol)

# Get adjacency matrix and edges from the graph
adj_matrix = get_adjacency_matrix(G)
edges = get_edges(adj_matrix)

# Initialize the FragmentationTree with the molecule
fragment_tree = FragmentationTree(mol)

# Set the depth of fragmentation
depth = 2  # Adjust depth as needed for your exploration

# Build the fragmentation tree
frag_tree = fragment_tree.build_fragmentation_tree(mol, edges, depth=depth)
```

```python
# Display the fragmentation tree
print("Fragmentation Tree:")
print(frag_tree)
```
